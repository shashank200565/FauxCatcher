```python
from transformers import AutoImageProcessor, AutoModelForImageClassification
from PIL import Image
import torch
```

```python
processor = AutoImageProcessor.from_pretrained("umm-maybe/ai-image-detector")
model = AutoModelForImageClassification.from_pretrained("umm-maybe/ai-image-detector")
```

```python
import requests
from io import BytesIO
from PIL import Image

url = "https://cdn.pixabay.com/photo/2024/04/15/07/55/ai-generated-8697226_1280.png"
image = Image.open(BytesIO(requests.get(url).content)).convert("RGB")
image.save("sample.jpg")
image_path = "sample.jpg"

print("✅ Image downloaded and saved as sample.jpg")
image.show()
```

```python
def detect_deepfake(image_path):
    image = Image.open(image_path).convert("RGB")
    inputs = processor(images=image, return_tensors="pt")
    outputs = model(**inputs)

    probs = torch.nn.functional.softmax(outputs.logits, dim=-1)
    labels = model.config.id2label

    confidence, label = torch.max(probs, 1)
    return labels[label.item()], float(confidence)
```

```python
label, confidence = detect_deepfake(image_path)
print(f"🧠 Deepfake detection result: {label} ({confidence*100:.2f}% confidence)")
```
